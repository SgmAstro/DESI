```python
import pylab             as pl
import astropy.io.fits   as fits
import matplotlib.pyplot as plt

from   astropy.table     import Table
```

```python
fpath = '/cosma/home/durham/dc-wils7/data/GAMA4/gama_gold_G9_ddp_n8_d0_6_lumfn.fits'

dat   = Table.read(fpath)
dat
```

```python
pl.semilogy(dat['MEDIAN_M'], dat['PHI_IVMAX'])
pl.semilogy(dat['MEDIAN_M'], dat['PHI_IVMAX_JK'], linestyle='--')
```

```python
pl.plot(dat['MEDIAN_M'], dat['PHI_IVMAX_ERROR'] / dat['PHI_IVMAX_ERROR_JK'])

pl.xlabel(r'$M$')
pl.ylabel(r'$\sigma \ / \ \sigma_{JK}$')

pl.axhline(1.0, c='k', lw=0.5)

pl.xlim(-24., -16.)
pl.ylim(0.3, 4.0)
```
