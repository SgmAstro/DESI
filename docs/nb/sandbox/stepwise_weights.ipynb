```python
fpath  = findfile('ddp', version='v2')

ddp    = Table.read(fpath)
ddp
```

```python
pl.plot(ddp['ZGAMA'], 1. / ddp['WEIGHT_STEPWISE'], marker=',', lw=0.0)

pl.xlabel('$z$', fontsize=16)
pl.ylabel('Inverse stepwise weight $\equiv \overline n(z)$', fontsize=16)
```

```python
fpath = findfile('vmax')
vmax  = Table.read(fpath)
vmax
```

```python
dM = 0.5
Ms = np.arange(-23., -17., dM)

dV = 0.025

for MM in Ms:
    isin = (vmax['MCOLOR_0P0'] > MM) & (vmax['MCOLOR_0P0'] < MM + dM)
    
    vv   = vmax[isin]    
    zmax = vmax['ZMAX'].max()
    vlim = vmax['VMAX'].max()
    
    nn   = vmax[vmax['ZGAMA'] < zmax]
    
    Mcnts, Mbins = np.histogram(vv['VZ'] / vv['VMAX'], bins=np.arange(-0.025, 1.025, dV))
    cnts,   bins = np.histogram(nn['VZ'] /       vlim, bins=np.arange(-0.025, 1.025, dV), weights=nn['WEIGHT_STEPWISE']) 

    pl.xlabel(r'$V/V_{\rm Max}$')
    pl.title(r'${:.3f} < $ MCOLOR_0P0 $ < {:.3f}$ to $z<{:.3f}$'.format(MM, MM+dM, zmax), fontsize=18)
    
    pl.axhline(dV, c='k', lw=0.5)
    
    toplot = [Mcnts, 1. * Mcnts / cnts]
    labels = ['Unweighted', 'Stepwise']
    
    for toplot, label in zip(toplot, labels):
        toplot  = np.array(toplot).astype(float)

        isin    = np.isfinite(toplot)
    
        toplot /= np.sum(toplot[isin])
    
        plt.step(Mbins[:-1][isin] + dV, toplot[isin], where='mid', lw=0.5, label=label)
    
    pl.legend(frameon=False, loc=4, fontsize=14)
    pl.show()
```
