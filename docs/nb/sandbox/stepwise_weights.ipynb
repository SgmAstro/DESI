```python
import os
import sys
import pylab as pl
import numpy as np

sys.path.append('/cosma/home/durham/dc-wils7/DESI/')

import pubplot

from   astropy.table import Table
from   findfile      import findfile
from   schechter     import named_schechter
```

```python
os.environ['GOLD_DIR']    = '/cosma5/data/durham/dc-wils7/GAMA4/'
os.environ['RANDOMS_DIR'] = '/cosma5/data/durham/dc-wils7/GAMA4/randoms/'
```

```python
survey = 'gama'
```

```python
stepwise          = Table.read('/cosma5/data/durham/dc-wils7/GAMA4/gama_gold_lumfn_step.fits')
stepwise
```

```python
Ms = np.arange(-23., -15., 0.01)
sc = named_schechter(Ms, named_type='TMR')

pl.plot(Ms, np.log10(sc), label='TMR', lw=2)

pl.plot(stepwise['Ms'] , -1.1 + np.log10(stepwise['PHI_STEP']), label='STEPWISE', alpha=0.75, lw=2)

pl.xlabel(r'$M_r^e - 5 \log_{10}|h|$')
pl.ylabel(r'$\log_{10} \left[ \frac{\phi_\delta(M)}{h^{-3} \ {\rm Mpc}^{3} \ {\rm Mag}^{}} \right] $')

pl.ylim(-4.25, -1.25)
pl.xlim(-23., -15.5)

pl.legend(frameon=False, loc=2)
pl.title(survey.upper())
```

```python
fpath  = findfile('ddp', version='v2')

ddp    = Table.read(fpath)
ddp
```

```python
pl.plot(ddp['ZGAMA'], 1. / ddp['WEIGHT_STEPWISE'], marker=',', lw=0.0)

pl.xlabel('$z$', fontsize=16)
pl.ylabel('Stepwise weight $\equiv 1 \ / \ \overline n(z)$', fontsize=16)
```

```python
fpath = findfile('vmax', version='v2')
vmax  = Table.read(fpath)

pl.plot(vmax['MCOLOR_0P0'], vmax['VMAX'] / vmax['VMAX'].max(), marker=',', lw=0.0, label='VMAX(MCOLOR_0P0)')

pl.xlim(-16., -26.)

pl.xlabel('Abs. Mag.')
pl.ylabel('VMAX')

pl.legend(frameon=False)
```
