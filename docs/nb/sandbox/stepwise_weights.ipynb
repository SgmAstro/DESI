```python
fpath = findfile('vmax')
# fpath = '/cosma5/data/durham/dc-moor2/GAMA4/gama_gold_ddp.fits'

vmax  = Table.read(fpath)

pl.plot(vmax['MCOLOR_0P0'], vmax['VMAX'] / vmax['VMAX'].max(), marker=',', lw=0.0, label='VMAX(MCOLOR_0P0)')
pl.plot(stepwise['M'], stepwise['VMAX'], c='m', label='EEP (DDPMALL_0P0)', lw=0.5)

pl.xlim(-16., -26.)

pl.xlabel('Abs. Mag.')
pl.ylabel('VMAX')

pl.legend(frameon=False)
```
