```python
import numpy as np
import pylab as pl

import matplotlib.pyplot as plt
```

```python
cthetas = np.arange(-1., 1., 1.e-4)
phis    = np.linspace(0.0, 2. * np.pi, len(cthetas))

thetas  = np.arccos(cthetas)

np.random.shuffle(phis)
```

```python
zs      = np.cos(thetas)
ys      = np.sin(thetas) * np.sin(phis)
xs      = np.sin(thetas) * np.cos(phis)
```

```python
pos    = np.c_[xs, ys, zs]
pos
```

```python
_ = pl.hist(pos[:,0], histtype='step', bins=np.arange(-1., 1., 0.1), label='x')
_ = pl.hist(pos[:,1], histtype='step', bins=np.arange(-1., 1., 0.1), label='y')
_ = pl.hist(pos[:,2], histtype='step', bins=np.arange(-1., 1., 0.1), label='z')

pl.legend(loc=3)
```

```python
morph = [np.count_nonzero(p > 0.0) for p in pos]
```

```python
fig   = plt.figure(figsize=(20,20))
ax    = fig.add_subplot(projection='3d')

sc    = ax.scatter(pos[::1, 0],\
                   pos[::1, 1],\
                   pos[::1, 2],\
                   marker='.',\
                   c=morph,\
                   lw=0.0)

# pl.colorbar(sc)
```
